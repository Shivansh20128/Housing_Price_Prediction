```python
import pandas as pd

df = pd.read_csv("housing.csv")
df.head()
```

```python
df_shuffled = df.sample(len(df),random_state=1)
df_shuffled
```

```python
pd.get_dummies(df_shuffled['ocean_proximity']).head()
```

```python
df_new = pd.concat([df_shuffled.drop('ocean_proximity', axis=1), pd.get_dummies(df_shuffled['ocean_proximity'])],axis=1)
df_new
```

```python
df_new = df_new[['longitude','latitude','housing_median_age','total_rooms','total_bedrooms','population','households','median_income','<1H OCEAN','INLAND' , 'ISLAND','NEAR BAY','NEAR OCEAN', 'median_house_value']]
df_new
```

```python
df_new = df_new.dropna()
```

```python
train_set, test_set, validation_set = df_new[:18000], df_new[18000:19217], df_new[19216:]
x_train, y_train = train_set.to_numpy()[:,:-1], train_set.to_numpy()[:,-1]
# x_train.shape
y_train.shape
```

```python
x_test, y_test = test_set.to_numpy()[:,:-1], test_set.to_numpy()[:,-1]
x_val, y_val = validation_set.to_numpy()[:,:-1], validation_set.to_numpy()[:,-1]
x_train.shape, y_train.shape, x_val.shape, y_val.shape, x_test.shape, y_test.shape
```

```python
from sklearn.preprocessing import StandardScaler
import numpy as np

scalar = StandardScaler().fit(x_train[:, :8])

def preProcessor(X):
    A = np.copy(X)
    A[:,:8] = scalar.transform(A[:,:8])
    return A

x_train = preProcessor(x_train)
x_test = preProcessor(x_test)
x_val = preProcessor(x_val)
```

```python
pd.DataFrame(x_train)[0].hist()
```

```python
from sklearn.metrics import mean_squared_error as mse
from sklearn.linear_model import LinearRegression

lm = LinearRegression().fit(x_train,y_train)
mse_val_train, mse_val_validation = mse(lm.predict(x_train), y_train, squared=False), mse(lm.predict(x_val), y_val, squared=False)
mse_val_train, mse_val_validation 
```

```python
from sklearn.neighbors import KNeighborsRegressor

knn = KNeighborsRegressor(n_neighbors=12).fit(x_train, y_train)
mse(knn.predict(x_train), y_train, squared=False), mse(knn.predict(x_val), y_val, squared=False) 
```

```python
from sklearn.ensemble import RandomForestRegressor

rfr = RandomForestRegressor(max_depth=10).fit(x_train, y_train)
mse(rfr.predict(x_train), y_train, squared=False), mse(rfr.predict(x_val), y_val, squared=False) 
```

```python
from sklearn.ensemble import GradientBoostingRegressor

gbr = GradientBoostingRegressor(n_estimators=250).fit(x_train, y_train)
mse(gbr.predict(x_train), y_train, squared=False), mse(gbr.predict(x_val), y_val, squared=False) 
```
