```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import tensorflow as tf
from tensorflow import keras
```

```python
# Getting the dataset and storing dividing it into training and testing

(X_train, y_train) , (X_test, y_test) = keras.datasets.mnist.load_data()
```

```python
# Scaling our dataset to get better results

X_test = X_test/255
X_train = X_train/255

X_train[0] #printing the training dataset
```

```python
# Showing a sample

plt.matshow(X_train[2])
```

```python
# Flattening the dataset to feed to the model
X_train_flattened = X_train.reshape(len(X_train), 28*28)
X_test_flattened = X_test.reshape(len(X_test), 28*28)

X_train_flattened.shape
```

```python
# This is a model with a single layer - The input layer and the output layer. There are no hidden layers.
#  10 means that we have 10 classes: 0 to 9. 

model = keras.Sequential([
    keras.layers.Dense(10, input_shape=(784,), activation='sigmoid')
])

model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

model.fit(X_train_flattened, y_train, epochs=5)
```

```python
model.evaluate(X_test_flattened, y_test)
# This will give the loss function value and the accuracy on the testing set.
```

```python
# now we are making predictions on the test set.
# Note that y_predicted does not store the labels for the test set. Infact, it stores the value of likelihood of each sample as an integer, for all samples.

y_predicted = model.predict(X_test_flattened)
```

```python
# Here we obtain the most likely predicted value for all samples, and store it.

predicted_labels = [np.argmax(i) for i in y_predicted]
predicted_labels
```

```python
# Now we make a confusion matrix to assess our model
confusion_matrix = tf.math.confusion_matrix(labels=y_test, predictions=predicted_labels)
confusion_matrix
```

```python
# To judge our model better, we will make a heatmap

import seaborn as sn
plt.figure(figsize = (10,7))
sn.heatmap(confusion_matrix, annot=True, fmt='d')
plt.xlabel('Predicted')
plt.ylabel('Truth')
```

Now, we will make a new model, more complex than before. This will have a single hidden layer.

```python
# Flattening the dataset using keras

model = keras.Sequential([
    keras.layers.Flatten(input_shape=(28,28)),
    keras.layers.Dense(100,activation='relu'),
    keras.layers.Dense(10,  activation='sigmoid')
])

model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

model.fit(X_train, y_train, epochs=5)
```

```python
# Making predictions on test set

predictions_new = model.predict(X_test)
```

```python
# Getting labels

new_predicted_labels = [np.argmax(i) for i in predictions_new]
new_predicted_labels
```

```python
# Confusion matrix

new_confusion_matrix = tf.math.confusion_matrix(labels=y_test, predictions=new_predicted_labels)
new_confusion_matrix
```
