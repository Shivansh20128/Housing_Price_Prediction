```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn.datasets
from sklearn.model_selection import train_test_split
```

```python
# loading the dataset
dataset = sklearn.datasets.load_breast_cancer()
```

```python
# storing data into a pandas dataframe
df = pd.DataFrame(dataset.data,columns=dataset.feature_names)
```

```python
# printing the dataframe
df
```

```python
# adding the target column to the dataframe
df["label"] = dataset.target
df
```

```python
# Extracting features and target from the prepared dataframe
features = df.drop(columns= "label", axis=1)
target_label = df["label"]
```

```python
features
```

```python
target_label
```

```python
# Scaling the dataset to get better results
from sklearn.preprocessing import StandardScaler
scalar = StandardScaler()
scaled_data = scalar.fit_transform(features)
scaled_data
```

```python
# Splitting the dataset
X_train, X_test, Y_train, Y_test = train_test_split(scaled_data, target_label, test_size = 0.2, random_state = 2)
```

```python
print("Total data: ", scaled_data.shape)
print("Train data: ", X_train.shape)
print("Test data: ", X_test.shape)
```

```python
import tensorflow as tf
from tensorflow import keras

# preparing neural network model using keras with 1 hidden layer
model = keras.Sequential([
    keras.layers.Flatten(input_shape = (30,)),
    keras.layers.Dense(20, activation = 'relu'),
    keras.layers.Dense(2, activation = "sigmoid")
])

# compiling the model
model.compile(optimizer = "adam", loss = "sparse_categorical_crossentropy", metrics = ["accuracy"])
```

```python
# training the model on train dataset
model_trained = model.fit(X_train, Y_train, validation_split = 0.1, epochs = 10)
```

```python
# Plotting the accuracy of model on training and validation set
plt.plot(model_trained.history["accuracy"])
plt.plot(model_trained.history["val_accuracy"])

plt.title("Model Accuracy")
plt.ylabel("accuracy")
plt.xlabel("epoch")

plt.legend(["training", "validation"], loc = "lower right")
```

```python
# Plotting the loss curve of model on training and validation sets
plt.plot(model_trained.history["loss"])
plt.plot(model_trained.history["val_loss"])

plt.title("Loss Curve")
plt.ylabel("loss")
plt.xlabel("epoch")

plt.legend(["training", "validation"], loc = "upper right")
```

```python
# Making predictions using the model
predictions = model_trained.model.predict(X_test)
predictions
```

```python
# getting the loss and accuracy on test set
loss, accuracy = model_trained.model.evaluate(X_test, Y_test)
accuracy
```

```python
# Converting the labels to integers
predcition_labels = [np.argmax(i) for i in predictions]
print(predcition_labels)
```

```python
# Now extract a single sample from the sataframe
sample = df.drop("label", axis = 1).iloc[0].values

# reshaping the sample according to our model
input_sample = sample.reshape(1, -1)

# standardize the sample
std_sample = scalar.transform(input_sample)
```

```python
# making predictions
prediction = model.predict(std_sample)
```

```python
# Getting predicted label
prediction_label = [np.argmax(prediction)]
print(prediction_label)

if(prediction_label[0] == 0):
    print("Malignant")
else:
    print("Benign.")
```
