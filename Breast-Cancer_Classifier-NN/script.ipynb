```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn.datasets
from sklearn.model_selection import train_test_split
```

```python
dataset = sklearn.datasets.load_breast_cancer()
```

```python
dataset
```

```python
df = pd.DataFrame(dataset.data,columns=dataset.feature_names)
```

```python
df
```

```python
df["label"] = dataset.target
df
```

```python
features = df.drop(columns= "label", axis=1)
target_label = df["label"]
```

```python
features
```

```python
target_label
```

```python
from sklearn.preprocessing import StandardScaler
scalar = StandardScaler()
scaled_data = scalar.fit_transform(features)
scaled_data
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(scaled_data, target_label, test_size = 0.2, random_state = 2)
```

```python
print("Total data: ", scaled_data.shape)
print("Train data: ", X_train.shape)
print("Test data: ", X_test.shape)
```

```python
import tensorflow as tf
# tf.random.set_seed(3)
from tensorflow import keras
```

```python
model = keras.Sequential([
    keras.layers.Flatten(input_shape = (30,)),
    keras.layers.Dense(20, activation = 'relu'),
    keras.layers.Dense(2, activation = "sigmoid")
])

model.compile(optimizer = "adam", loss = "sparse_categorical_crossentropy", metrics = ["accuracy"])
```

```python
model_trained = model.fit(X_train, Y_train, validation_split = 0.1, epochs = 10)
```

```python
plt.plot(model_trained.history["accuracy"])
plt.plot(model_trained.history["val_accuracy"])

plt.title("Model Accuracy")
plt.ylabel("accuracy")
plt.xlabel("epoch")

plt.legend(["training", "validation"], loc = "lower right")
```

```python
plt.plot(model_trained.history["loss"])
plt.plot(model_trained.history["val_loss"])

plt.title("Model Accuracy")
plt.ylabel("loss")
plt.xlabel("epoch")

plt.legend(["training", "validation"], loc = "upper right")
```

```python
loss, accuracy = model.evaluate(X_test, Y_test)
accuracy
```

```python
predictions = model.predict(X_test)
predictions
```

```python
predcition_labels = [np.argmax(i) for i in predictions]
print(predcition_labels)
```

```python
sample = df.drop("label", axis = 1).iloc[0].values
input_data = (sample)
print(input_data)

# reshape the array as we are predicting for one instance
input_data_reshaped = input_data.reshape(1, -1)

# standardize the input data as model is trained as such
std_data = scalar.transform(input_data_reshaped)
print(std_data)

prediction = model.predict(std_data)
```

```python
predictin_label = [np.argmax(prediction)]
print(predictin_label)

if(predictin_label[0] == 0):
    print("The tumor is Malignant.")
else:
    print("The tumor is Benign.")
```
